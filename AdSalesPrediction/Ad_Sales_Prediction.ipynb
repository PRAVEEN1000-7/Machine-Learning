### Advertisement Sales prediction using Logistic Regression

## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Upload dataset :

```python
url="https://raw.githubusercontent.com/PRAVEEN1000-7/Machine-Learning/refs/heads/main/AdSalesPrediction/Ad_Sales_dataset.csv"
dataset = pd.read_csv(url)
dataset
```

# Segregate dataset into X (input) & Y (output)

```python
X = dataset.iloc[:,:-1].values
X
```

```python
Y = dataset.iloc[:,-1].values
Y
```

## Splitting the dataset for training & testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

## Feature Scaling

we scale our data to make all the features contribute equally to the result

* Fit_Tranform : fit method is calculating the mean and variance of each of the feature present in our data.

* Transform : transform method is transforming all the features using the respective mean and variance.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train,X_test)
```

## Training

```python
from sklearn.linear_model import  LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(X_train,Y_train)
```

## Testing

```python
Y_predict = model.predict(X_test)
print("actual output vs predicted output")
print(np.column_stack((Y_predict,Y_test)))
```

## Accuracy

```python
from sklearn.metrics import accuracy_score
print(f"Accuracy of the Model: {format(accuracy_score(Y_test,Y_predict)*100)} %")
```

## Prediction for whether a new customer will buy the product or not.

```python
try :
  Age = int(input("Enter New Customer Age: "))
  Salary = float(input("Enter New Customer Salary: "))
  newCust = [[Age,Salary]]
  result = model.predict(sc.transform(newCust))

  if result == 1:
    print("The customer is likely to make a purchase based on the given data and predictions.")
  else:
     print("The customer is unlikely to make a purchase based on the given data and predictions.")
except :
  print("Invalid Input,please try again")
```
